```python
# SO CI
from mplsoccer import Radar, FontManager, grid
import matplotlib.pyplot as plt

# plt.rcParams['figure.dpi'] = 1000  # For inline display
# plt.rcParams['savefig.dpi'] = 1000 # For saved figures

# parameter names of the statistics we want to show
params = ["TM-NLL", "T-NLL", "M-NLL", "RMSE", "F1"]

# The lower and upper boundaries for the statistics
low =  [2.1, 0.6, 1.4, 1.0, 0.2]
high = [2.5, 1.0, 1.8, 1.4, 0.32]

# Add anything to this list where having a lower number is better
# this flips the statistic
lower_is_better = ["TM-NLL", "T-NLL", "M-NLL", "RMSE"]

radar = Radar(params, low, high,
              lower_is_better=lower_is_better,
              # whether to round any of the labels to integers instead of decimal places
              round_int=[False]*len(params),
              num_rings=4,  # the number of concentric circles (excluding center circle)
              # if the ring_width is more than the center_circle_radius then
              # the center circle radius will be wider than the width of the concentric circles
              ring_width=1, center_circle_radius=1)

itpp_values = [2.106, 0.636, 1.47, 1.007, 0.313]
wo_values = [2.228, 0.715, 1.514, 1.009, 0.313]

fig, axs = grid(figheight=8, grid_height=0.915, title_height=0.06, # endnote_height=0.025,
                title_space=0, endnote_space=0, grid_key='radar', axis=False)
ax = axs['radar']
radar.setup_axis(ax=ax)
rings_inner = radar.draw_circles(ax=ax, facecolor='white', edgecolor='black')

radar1, vertices1 = radar.draw_radar_solid(itpp_values, ax=ax,
                                           kwargs={'facecolor': '#aa65b2',
                                                   'alpha': 0.6,
                                                   'edgecolor': '#216352',
                                                   'lw': 2})

radar2, vertices2 = radar.draw_radar_solid(wo_values, ax=ax,
                                           kwargs={'facecolor': '#66d8ba',
                                                   'alpha': 0.6,
                                                   'edgecolor': '#216352',
                                                   'lw': 2})

lines = radar.spoke(ax=ax, color='#a6a4a1', linestyle='--', zorder=2)

ax.scatter(vertices1[:, 0], vertices1[:, 1],
           c='#aa65b2', edgecolors='#502a54', marker='o', s=50, zorder=2)
ax.scatter(vertices2[:, 0], vertices2[:, 1],
           c='#66d8ba', edgecolors='#216352', marker='o', s=50, zorder=2)



range_labels = radar.draw_range_labels(ax=ax, fontsize=18)
param_labels = radar.draw_param_labels(ax=ax, fontsize=20)


# title1_text = axs['title'].text(0.01, 0.65, 'ITPP', fontsize=32, color='#aa65b2',
#                                 ha='left', va='center')
# title3_text = axs['title'].text(0.99, 0.65, 'ITPP w/o CI', fontsize=32,
#                                 ha='right', va='center', color='#66d8ba')
```
